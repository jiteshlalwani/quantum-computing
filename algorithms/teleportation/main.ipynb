```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit_textbook.tools import array_to_latex
from qiskit.visualization import plot_histogram
```

qc_tel = QuantumCircuit(3, 3)
initial_state = [0, 1]
qc_tel.initialize(initial_state, 0)
qc_tel.draw(output='mpl')

```python
qc_tel.x(0)
qc_tel.barrier()
qc_tel.draw(output='mpl')
```

```python
qc_tel.h(1)
qc_tel.cx(1, 2)
qc_tel.barrier()
qc_tel.draw(output='mpl')
```

```python
qc_tel.cx(0, 1)
qc_tel.h(0)
qc_tel.barrier()
qc_tel.draw(output='mpl')
```

```python
qc_tel.measure(0, 0)
qc_tel.measure(1, 1)
qc_tel.barrier()
qc_tel.draw(output='mpl')
```

```python
qc_tel.cx(1, 2)
qc_tel.cz(0, 2)
qc_tel.draw(output='mpl')
```

```python
qc_tel.measure(2, 2)
backend = Aer.get_backend('qasm_simulator')
result = execute(qc_tel, backend).result()
counts = result.get_counts()
plot_histogram(counts)
```
