```python
from qiskit import QuantumCircuit, Aer, execute
```

```python
secret_num = "11011001"

len_num = len(secret_num)

qc_vaz = QuantumCircuit(len_num + 1, len_num)

for i in range(len_num):
    qc_vaz.h(i)
    
qc_vaz.x(len_num)
qc_vaz.h(len_num)

qc_vaz.barrier()
qc_vaz.draw(output='mpl')
```

```python
rev_secret_num = reversed(secret_num)
for count, item in enumerate(rev_secret_num):
    if item == '1':
        qc_vaz.cx(count, len_num)
qc_vaz.barrier()
qc_vaz.draw(output='mpl')      
```

```python
for i in range(len_num):
    qc_vaz.h(i)
qc_vaz.barrier()
qc_vaz.draw(output='mpl')
```

```python
for i in range(len_num):
    qc_vaz.measure(i, i)
qc_vaz.draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
result = execute(qc_vaz, backend, shots = 1).result()
print(result.get_counts())
```
